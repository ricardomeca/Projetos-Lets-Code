```python
# Projeto do Módulo 2 (Professor: Rafael)	
# Aula 1	Strings
# Aula 2	Arquivos
# Aula 3	Tratamento de Exceção
# Aula 4	Parâmetros de Funções
# Aula 5	Compreensão de listas

#OBS: Nome do arquivo csv: lista_nomes.csv enviado juntamente com o projeto

import csv
def tela_opcoes() -> str: # função que abre tela de opções  
    print('''
 ===============================================
     Seja bem vindo ao MECA LETS COLDEPLAY  
 ===============================================
 1. Cadastrar músicos
 2. Buscar músicos
 3. Modificar músicos
 4. Montar bandas
 5. Encerrar o programa
 ===============================================
  Digite uma das opções acima e pressione ENTER
 ==============================================='''
    )
    opcao_escolhida = input()
    return opcao_escolhida

def menu_variaveis(opcao_escolhida): # função que chama as opções escolhidas no menu. Conteúdos das Aulas 1 e 4 + ternários
    dicionario = {'1': 'cadastrar','2': 'buscar','3': 'modificar','4': 'montar'}
    while opcao_escolhida != '5':
        if opcao_escolhida not in dicionario.keys():
            print(f"{'OPÇÃO INVÁLIDA!':^47}",'\n','='*47)
            opcao_escolhida = tela_opcoes()
            continue
       
        if opcao_escolhida == '1':
            cadastrar_musicos()
               
        if opcao_escolhida == '2': # Aula 4 de *args
            lista_busca = coleta_dados_pesquisa()
            busca_usuario(*lista_busca)
        
        if opcao_escolhida == '3':
            modifica_usuario()

        if opcao_escolhida == '4':
            
            numero = 'a'
            while type(numero) != int:
                try:
                    numero = int(input('Digite a quantidade de músicos da banda:'))
                    if numero <= 0:
                        print('Valor deve ser positivo! Tente novamente.')
                        numero = 'a'
                except ValueError:
                    print('Valor inválido!')
            
            genero = input('Digite o gênero desejado:')
            lista_instrumentos = []
            for musicos in range(int(numero)):
                instrumento = input(f'Digite o instrumento do {musicos+1}° músico:')
                lista_instrumentos.append(instrumento)
            print('\nLista de instrumentos: ',lista_instrumentos)
            lista_email_instrumentos = monta_lista_email_instrumentos_banda(genero)
            if lista_email_instrumentos == []:
                print('='*47)
                print(f'Não há músicos cadastrados do gênero: {genero}')

            lista_intermediaria =[]

            cont = 0
            for cada_instrumento in lista_instrumentos:
                for linha in lista_email_instrumentos: 
                    for cada_elemento in linha[1]:
                        if cada_elemento == cada_instrumento:
                            lista_intermediaria.append([])
                            lista_intermediaria[cont].append(linha[0])
                            lista_intermediaria[cont].append(cada_instrumento)
                            cont+=1   
            
            for linha in lista_intermediaria:
                linha[0]=linha[0]+': '+linha[1]
                       
                        
            lista_final = []
            contador = 0 
            while contador < len(lista_instrumentos):
                lista_final.append([])
                for linha in lista_intermediaria:
                    if linha[1] == lista_instrumentos[contador]:
                        lista_final[contador].append(linha[0]) 
                contador+=1
            
            numero = len(lista_instrumentos)   
            lista_combinada = combinacoes(numero, lista_final)
            
            lista_auxiliar=[]

            for linha in lista_combinada:
                string = ' '.join(linha)
                lista_auxiliar.append(string.split())
            
            listaduplicado = [] #removendo duplicados
            lista_indice_remocao = []

            cont = 0
            for cada in lista_auxiliar:
                for i,x in enumerate(cada):
                    if i != cada.index(x):
                        listaduplicado.append(f'problema na linha: {cont}')
                        lista_indice_remocao.append(cont)
                cont+=1     

            lista_indice_remocao.reverse() #índice dos elementos a serem retirados (de trás prá frente)
            for cada in lista_indice_remocao:
                lista_combinada.pop(cada)
            cont = 1
            print('='*47)
          
            if len(lista_combinada) == 0:
                print('Impossível de formar um grupo com os dados fornecidos!')
           
            for linha in lista_combinada:
                print(f'{cont}° grupo formado :{linha}')
                cont+=1
            print('='*47)
                
        opcao_escolhida = tela_opcoes()
    else:
        print('','='*47,f"\n{'Fim do MECA LETS COLDEPLAY':^47}",'\n','='*47)

def coletar_nome() -> str: # função que coleta o nome do usuário. Conteúdos da Aula 1 + ternário
    valida = False
    while not valida:  
        nome = input('Digite o nome do músico:').upper().strip()
        cont = 0
        cont_letras = 0
        for caracter in nome:
            if not (caracter.isalpha() or caracter.isspace()):
                cont +=1
            if caracter.isalpha():  
                cont_letras +=1
        valida = True if cont == 0 and cont_letras > 0 else print('Nome Inválido! Tente Novamente!')
    return nome

def coletar_email() -> str: # função que coleta o e-mail do usuário. Conteúdos da Aula 1 + ternário  
    valida = False
    while not valida or cont_arroba != 1 or len(email) <= 5: #(email com no mínimo 6 caracteres)  
        email = input('Digite o e-mail do músico:').lower()
        if len(email) <= 5:
            print ('O e-mail digitado não corresponde a um e-mail válido! Digite no mínimo 6 caracteres!') 
        cont = 0
        cont_arroba = 0
        for caracter in email:
            if not ((caracter.isalnum() or caracter == '_' or caracter == '.')):
                if caracter == '@':
                    cont_arroba+= 1
                else:
                    cont+= 1
        if cont_arroba > 1:
            print (f'E-mail inválido. Você digitou {cont_arroba} vezes o caracter @ !')
        valida = True if cont == 0 else print('E-mail Inválido! Tente Novamente!')
        
    return email

def coletar_genero() -> str: #  função que coleta os gêneros do usuário. Conteúdos das Aulas 1 e 3 
    string_genero = '' # Criei string pois tinha feito lista e o csv retornava string
    quantidade = '' 
    while type(quantidade) != int:
        try:
            quantidade = int(input('Quantos gêneros musicais deseja inserir? '))
        except ValueError:
            print ('O valor digitado não é um valor inteiro positivo e não nulo!')
            quantidade = str(quantidade)
        except TypeError:
            print ('O valor digitado não é um valor inteiro positivo e não nulo!')
        else:
            for c in range(quantidade):
                string_genero += input(f'Digite o {c+1}° gênero musical: ').strip().lower()
                if c < quantidade-1:
                    string_genero += ','
    return string_genero

def coletar_instrumentos() -> str: # função que coleta os e-mail do usuário. Conteúdos das Aulas 1, 2 e 3 
    string_instrumentos = '' # Criei string pois tinha feito lista e o csv retornava string
    quantidade = '' 
    while type(quantidade) != int:
        try:
            quantidade = int(input('Quantos instrumentos musicais deseja inserir? '))
        except ValueError:
            print ('O valor digitado não é um valor inteiro positivo e não nulo!')
            quantidade = str(quantidade)
        except TypeError:
            print ('O valor digitado não é um valor inteiro positivo e não nulo!')
        else:
            for c in range(quantidade):
                string_instrumentos += input(f'Digite o {c+1}° instrumento musical: ').strip().lower()
                if c < quantidade-1:
                    string_instrumentos += ','

    return string_instrumentos

def cadastrar_musicos() -> None: # função que cadastra os músicos. Conteúdos das Aulas 2 e 3
    nome = coletar_nome()
    email = coletar_email()
    string_genero = coletar_genero()
    string_instrumentos = coletar_instrumentos()
    cadastro = [[nome, email, string_genero, string_instrumentos]]

    try:           
        with open("lista_nomes.csv", "r") as documento:
            lista_final = list(csv.reader(documento, delimiter = ';', lineterminator='\n'))
            valida_email = True
            for cada_linha in lista_final:
                if cadastro[0][1] in cada_linha[1]:
                    valida_email = False  
            if valida_email:
                with open("lista_nomes.csv", "a") as documento:
                    regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
                    regra.writerows(cadastro)
                print (f'\nMúsico {nome} cadastrado com sucesso!!!')
            else:
                print('Este e-mail já existe. Músico não pode ser cadastrado.')
    except FileNotFoundError:
        resposta = 0
        while resposta != 'S' and resposta != 'N':
            resposta = input('Arquivo não encontrado. Deseja abrir um novo arquivo lista_nomes.csv? S/N').upper()
        if resposta == 'S':
            with open("lista_nomes.csv", "w") as documento:
                regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
                regra.writerows('')

def coleta_dados_pesquisa() -> list: # função que retorna uma lista com dados da pesquisa. Conteúdo: ternário
    
            escolha = 0
            while escolha != 'S' and escolha != 'N':
                escolha = input(' Deseja inserir o nome do músico? S/N ').upper()
            nome = coletar_nome() if escolha == 'S' else '' 
            
            escolha = 0
            while escolha != 'S' and escolha != 'N':
                escolha = input(' Deseja inserir o e-mail do músico? S/N ').upper()
            email = coletar_email() if escolha == 'S' else ''
            
            escolha = 0
            while escolha != 'S' and escolha != 'N':
                escolha = input(' Deseja inserir o gênero musical? S/N ').upper()
            genero = input(' Digite o gênero desejado:').lower()  if escolha == 'S' else ''
            
            escolha = 0
            while escolha != 'S' and escolha != 'N':
                escolha = input(' Deseja inserir o instrumento musical? S/N ').upper()
            instrumento = input(' Digite o instrumento desejado:').lower() if escolha == 'S' else ''
            
            print(f' ******** Dados coletados para a busca *********\n')
            print(' Nome: usuário optou em não digitar o nome na pesquisa.') if nome == '' else print(f' Nome: {nome}')
            print(' E-mail: usuário optou em não digitar o e-mail na pesquisa.') if email == '' else print(f' E-mail: {email}')
            print(' Gênero: usuário optou em não digitar o gênero na pesquisa.') if genero == '' else print(f' Gênero: {genero}')   
            print(' Instrumento: usuário optou em não digitar o instrumento na pesquisa.') if instrumento == '' else print(f' Instrumento: {instrumento}')
            print('','='*47)    

            opcao = 0
            while opcao != '1' and opcao != '2':
                opcao = input('''
    Escolha o tipo de busca:

    1. Resultados que batem com toda a lista digitada.
    2. Resultados que batem com pelo menos um elemento da lista digitada.
    
    '''
                 )

            lista_busca = [nome, email, genero, instrumento, opcao]
            return lista_busca

def busca_usuario(*lista_busca) -> None: # função que exibe a pesquisa solicitada. Conteúdos das Aulas 1,2,3 e 4

    try:
        with open("lista_nomes.csv", "r") as documento:
            lista_nomes = list(csv.reader(documento, delimiter = ';', lineterminator='\n'))
    except FileNotFoundError:
        print(' Não foi possível abrir o arquivo.')
        
    else:
        nome = lista_busca[0].upper()
        email = lista_busca[1].lower()
        genero = lista_busca[2].lower()
        instrumento = lista_busca[3].lower()
        opcao = lista_busca[4]
        
        for cada_linha in lista_nomes: # transformando as strings das colunas 2 e 3 em listas
            cada_linha[2] = cada_linha[2].split(',')
            cada_linha[3] = cada_linha[3].split(',')
                        
        if opcao == '1': # (nome email genero instrumento): 0(não forneceu) 1(forneceu)
            lista = []
            if nome == '' and email == '' and genero == '' and instrumento == '': #0000 (não forneceu nada)
                lista.append(' ')
                print(f"\n{' Não há usuários cadastrados!':^47}")
            
            if nome == '' and email == '' and genero == '' and instrumento != '': #0001 (só forneceu instrumento)
                for linha in lista_nomes: 
                    for cada_elemento in linha[3]:
                        if cada_elemento == instrumento:
                            lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1  
            
            if nome == '' and email == '' and genero != '' and instrumento == '': #0010 (só forneceu gênero)
                for linha in lista_nomes: 
                    for cada_elemento in linha[2]:
                        if cada_elemento == genero:
                            lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1      
            if nome == '' and email == '' and genero != '' and instrumento != '': #0011 (gênero e instumento)
                for linha in lista_nomes: 
                    for cada_elemento in linha[2]:
                        if cada_elemento == genero:
                            for elemento in linha[3]:
                                if elemento == instrumento:    
                                        lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1  

            if nome == '' and email != '' and genero == '' and instrumento == '': #0100 (só forneceu e-mail)
                for linha in lista_nomes: 
                    if linha[1] == email:
                        lista.append(linha)
                        print(f' Dados do usuário encontrado: {linha}')

            if nome == '' and email != '' and genero == '' and instrumento != '': #0101 (e-mail e instrumento)
                for linha in lista_nomes: 
                    if linha[1] == email:
                        for cada_elemento in linha[3]:
                            if cada_elemento == instrumento:
                                lista.append(linha)
                                print(f' Dados do usuário encontrado: {linha}')

            if nome == '' and email != '' and genero != '' and instrumento == '': #0110 (e-mail e gênero)
                for linha in lista_nomes: 
                    if linha[1] == email:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                lista.append(linha)
                                print(f' Dados do usuário encontrado: {linha}')

            if nome == '' and email != '' and genero != '' and instrumento != '': #0111 (e-mail, gênero e instrumento)
                for linha in lista_nomes: 
                    if linha[1] == email:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                for elemento in linha[3]:
                                    if elemento == instrumento:    
                                        lista.append(linha)
                                        print(f' Dados do usuário encontrado: {linha}')
            
            if nome != '' and email == '' and genero == '' and instrumento == '': #1000 (só forneceu o nome)
                for linha in lista_nomes: 
                    if linha[0] == nome:
                        lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1
            
            if nome != '' and email == '' and genero == '' and instrumento != '': #1001 (nome e instrumento)
                for linha in lista_nomes: 
                    if linha[0] == nome:
                        for cada_elemento in linha[3]:
                            if cada_elemento == instrumento:
                                lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1
                       
            if nome != '' and email == '' and genero != '' and instrumento == '': #1010 (nome e gênero)
                for linha in lista_nomes: 
                    if linha[0] == nome:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1     
            
            if nome != '' and email == '' and genero != '' and instrumento != '': #1011 (nome, gênero e instrumento)
                for linha in lista_nomes: 
                    if linha[0] == nome:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                for elemento in linha[3]:
                                    if elemento == instrumento:    
                                        lista.append(linha)
                cont = 1
                for cada_linha in lista:
                    print(f' Dados do {cont}° encontro: {cada_linha}')
                    cont+=1  
            
            if nome != '' and email != '' and genero == '' and instrumento == '': #1100 (nome e e-mail)
                for linha in lista_nomes: 
                    if linha[0] == nome and linha[1] == email:
                        lista.append(linha)
                        print(f' Dados do usuário encontrado: {linha}')
                            
            if nome != '' and email != '' and genero == '' and instrumento != '': #1101 (nome, e-mail e instrumento)
                for linha in lista_nomes: 
                    if linha[0] == nome and linha[1] == email:
                        for cada_elemento in linha[3]:
                            if cada_elemento == instrumento:
                                lista.append(linha)
                                print(f' Dados do usuário encontrado: {linha}')  
                
            if nome != '' and email != '' and genero != '' and instrumento == '': #1110 (nome, e-mail e gênero)
                for linha in lista_nomes: 
                    if linha[0] == nome and linha[1] == email:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                lista.append(linha)
                                print(f' Dados do usuário encontrado: {linha}')                 
                    
            if nome != '' and email != '' and genero != '' and instrumento != '': #1111 (nome, e-mail, gênero e instrumento)
                for linha in lista_nomes: 
                    if linha[0] == nome and linha[1] == email:
                        for cada_elemento in linha[2]:
                            if cada_elemento == genero:
                                for elemento in linha[3]:
                                    if elemento == instrumento:    
                                        lista.append(linha)
                                        print(f' Dados do usuário encontrado: {linha}')
                        
            if len(lista) == 0:
                print(f'\n{" Não há músicos cadastrados.":^47}') 
        
        if opcao == '2': # Contendo pelo menos 1
            lista = []
            for linha in lista_nomes:
                if linha[0] == nome:
                    lista.append(linha)
                elif linha[1] == email:
                    lista.append(linha)
                elif True:
                    for cada_elemento in linha[2]:
                        peguei = False
                        if cada_elemento == genero:
                            lista.append(linha)
                            peguei = True
                    if not peguei:
                        for cada_elemento in linha[3]:
                            if cada_elemento == instrumento:
                                lista.append(linha)
            cont = 1
            for cada_linha in lista:
                print(f' Dados do {cont}° encontro: {cada_linha}')
                cont+=1 
            if len(lista) == 0:
                print(f'\n{" Não há músicos cadastrados.":^47}')

def escolha_adicionar_remover() -> str: # função que abre tela de opções. Conteúdo da aula 4
    escolha = 0
    while escolha != '1' and escolha != '2' and escolha != '3' and escolha != '4':  
        escolha = input('''
    O que deseja fazer?
        
    1-> Remover um gênero;
    2-> Adicionar um gênero;
    3-> Remover um instrumento;
    4-> Adicionar um instrumento. 

    '''
        )
        if escolha != '1' and escolha != '2' and escolha != '3' and escolha != '4':
            print(f'\n{" Valor Inválido!!! Tente novamente!"}')
        if escolha == '1':
            texto_escolha = input(' Digite o gênero que deseja remover: ')
        if escolha == '2':
            texto_escolha = input(' Digite o gênero que deseja adicionar: ')
        if escolha == '3':
            texto_escolha = input(' Digite o instrumento que deseja remover: ')
            return escolha,texto_escolha
        if escolha == '4':
            texto_escolha = input(' Digite o instrumento que deseja adicionar: ')
    
    return escolha,texto_escolha

def modifica_usuario() -> None: # função que modifica o usuário.

    try:    
        with open("lista_nomes.csv", "r") as documento:
            lista_nomes = list(csv.reader(documento, delimiter = ';', lineterminator='\n'))

            for cada_linha in lista_nomes: # transformando as strings das colunas 2 e 3 em listas
                cada_linha[2] = cada_linha[2].split(',')
                cada_linha[3] = cada_linha[3].split(',')
            achou = False
            email = coletar_email()
            
            for linha in lista_nomes:
                if linha[1] == email:
                    achou = True
                    print('\n Músico encontrado: ', linha)
            if not achou:
                print(' E-mail não consta no cadastro! Você será redirecionado para o menu inicial.')
                tela_opcoes()

            escolha, texto_escolha = escolha_adicionar_remover()  
            if escolha == '1':
                escolha_usuario_1_remover_genero(lista_nomes,texto_escolha, email)
                    
            if escolha == '2':
                escolha_usuario_2_adicionar_genero(lista_nomes,texto_escolha, email)
                        
            if escolha == '3':
                escolha_usuario_3_remover_instrumento(lista_nomes,texto_escolha, email)
                
            if escolha == '4':
                escolha_usuario_4_adicionar_instrumento(lista_nomes,texto_escolha, email)
    
    except FileNotFoundError: 
        print(' Arquivo não encontrado.\n Você será direcionado para o menu principal.')
        tela_opcoes()

def escolha_usuario_1_remover_genero(lista_nomes:list, texto_escolha:str, email:str) -> None: # função de remover gênero. Conteúdos das aulas 1 e 2.
    removeu = False
    for linha in lista_nomes:
        if linha[1] == email:
            for cada_elemento in linha[2]:
                if texto_escolha == cada_elemento:
                    linha[2].remove(texto_escolha)
                    removeu = True
                    print (f' Gênero {texto_escolha} removido com sucesso!')
                    break
        
    if not removeu:       
        print (f' Impossível remover o gênero {texto_escolha} pois o usuário {email} não possui esse gênero!')
    
    for cada_linha in lista_nomes: # transformando listas em strings
        cada_linha[2] =','.join(cada_linha[2])
        cada_linha[3] =','.join(cada_linha[3])
            
    with open("lista_nomes.csv", "w") as documento:
        regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
        regra.writerows(lista_nomes)

def escolha_usuario_2_adicionar_genero(lista_nomes:list, texto_escolha:str, email:str) -> None: # função de adicionar gênero. Conteúdos das aulas 1 e 2.

    for linha in lista_nomes:
        if linha[1] == email:
            if texto_escolha not in linha[2]:
                linha[2].append(texto_escolha)
                print (f' Gênero {texto_escolha} adicionado com sucesso!')
            else:
                print (f' Gênero {texto_escolha} já estava cadastrado no músico!')
            
            print(f' Gêneros atuais do músico:\n {linha[2]}')
    
    for cada_linha in lista_nomes: # transformando listas em strings e resolvendo o problema de lista vazia
        if cada_linha[2][0] == '':
            cada_linha[2].pop(0)
        cada_linha[2] = ','.join(cada_linha[2])
        if cada_linha[3][0] == '':
            cada_linha[3].pop(0) 
        cada_linha[3] = ','.join(cada_linha[3])
            
    with open("lista_nomes.csv", "w") as documento:
        regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
        regra.writerows(lista_nomes)

def escolha_usuario_3_remover_instrumento(lista_nomes:list, texto_escolha:str, email:str) -> None: # função de remover instrumento. Conteúdos das aulas 1 e 2.       
    removeu = False
    for linha in lista_nomes:
        if linha[1] == email:
            for cada_elemento in linha[3]:
                if texto_escolha == cada_elemento:
                    linha[3].remove(texto_escolha)
                    removeu = True
                    print (f' Instrumento {texto_escolha} removido com sucesso!')
                    break
        
    if not removeu:       
        print (f' Impossível remover o instrumento {texto_escolha} pois o usuário {email} não possui esse instrumento!')
    
    for cada_linha in lista_nomes: # transformando listas em strings
        cada_linha[2] =','.join(cada_linha[2])
        cada_linha[3] =','.join(cada_linha[3])
            
    with open("lista_nomes.csv", "w") as documento:
        regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
        regra.writerows(lista_nomes)

def escolha_usuario_4_adicionar_instrumento(lista_nomes:list, texto_escolha:str, email:str) -> None: # função de adicionar instrumento. Conteúdos das aulas 1 e 2.
    for linha in lista_nomes:
            if linha[1] == email:
                if texto_escolha not in linha[3]:
                    linha[3].append(texto_escolha)
                    print (f' Instrumento {texto_escolha} adicionado com sucesso!')
                else:
                    print (f' Instrumento {texto_escolha} já estava cadastrado no músico!')
                  
                print(f' Instrumentos atuais do músico:\n {linha[3]}')

    for cada_linha in lista_nomes: # transformando listas em strings e resolvendo o problema de lista vazia
        if cada_linha[2][0] == '':
            cada_linha[2].pop(0)
        cada_linha[2] = ','.join(cada_linha[2])
        if cada_linha[3][0] == '':
            cada_linha[3].pop(0) 
        cada_linha[3] = ','.join(cada_linha[3])
                        
    with open("lista_nomes.csv", "w") as documento:
        regra = csv.writer(documento, delimiter = ';', lineterminator='\n')
        regra.writerows(lista_nomes)   

def monta_lista_email_instrumentos_banda(genero:str) -> list: # função que monta uma lista solicitada na pesquisa do usuário. Conteúdo das aulas 1, 2 e 3.

    try: 
        with open("lista_nomes.csv", "r") as documento:
                lista_nomes = list(csv.reader(documento, delimiter = ';', lineterminator='\n'))
        
        for cada_linha in lista_nomes: # transformando as strings das colunas 2 e 3 em listas
            cada_linha[2] = cada_linha[2].split(',')
            cada_linha[3] = cada_linha[3].split(',')

        lista_email_instrumentos =[]    
        cont = 0
        for linha in lista_nomes: 
                for cada_elemento in linha[2]:
                    if cada_elemento == genero:
                        lista_email_instrumentos.append([])
                        lista_email_instrumentos[cont].append(linha[1])
                        lista_email_instrumentos[cont].append(linha[3])
                        cont+=1
        return lista_email_instrumentos
    except FileNotFoundError:
        print(' Arquivo não encontrado.\n Você será direcionado para o menu principal.')
        tela_opcoes()  

def combinacoes (n:int, lista:list, lista_aux:list = []) -> list: # Conteúdo da aula 4 - Solução com a ajuda do professor Rafael    
    if len(lista_aux) == n:
        return [lista_aux]
    
    lista_combinacoes = []
    for linha in lista[0]:
        lista_copia = lista_aux [:]
        lista_copia.append(linha)
        lista_combinacoes+= combinacoes(n, lista[1:], lista_copia) # lista de combinacoes cada elemento tem tamanho n
    return lista_combinacoes


opcao_escolhida = tela_opcoes()
menu_variaveis(opcao_escolhida)
```
